```python
import kagglehub
import os
import pandas as pd

# Download latest version
path = kagglehub.dataset_download("kandij/mall-customers")
cpath = os.path.join(path, "Mall_Customers.csv")
df = pd.read_csv(cpath)
df.head()
```

```python
x = df.iloc[:, [3,4]].values
```

```python
x
```

```python
#the elbow method
from sklearn.cluster import KMeans
error = []
for i in range (2 , 10):
  k = KMeans(n_clusters=i, init='k-means++')
  k.fit(x)
  error.append(k.inertia_)
```

```python
import matplotlib.pyplot as plt
plt.plot(range(2,10), error)
plt.title('the elbow method')
plt.xlabel('number of clusters')
plt.ylabel('error')
plt.show()
```

```python
k = KMeans(n_clusters=5, init='k-means++')
k.fit(x)
```

```python
y = k.predict(x)
y
```

```python
plt.scatter(x[y==0, 0], x[y==0, 1], s=100, c='red', label='cluster 1')
plt.scatter(x[y==1, 0], x[y==1, 1], s=100, c='blue', label='cluster 2')
plt.scatter(x[y==2, 0], x[y==2, 1], s=100, c='green', label='cluster 3')
plt.scatter(x[y==3, 0], x[y==3, 1], s=100, c='cyan', label='cluster 4')
plt.scatter(x[y==4, 0], x[y==4, 1], s=100, c='magenta', label='cluster 5')
plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```

RECOMMENDATION SYSTEM FOR GROCERY STORE.

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
df = pd.read_csv('/content/groceries.csv')
```

```python
txns = df.apply(lambda row: row.dropna().tolist(),axis = 1).tolist()
txns
```

```python
t = TransactionEncoder()
te_ary = t.fit(txns).transform(txns)
dt_encoded = pd.DataFrame(te_ary, columns = t.columns_)
dt_encoded
```

```python
frequent_items = apriori(dt_encoded, min_support = 0.03, use_colnames= True)
frequent_items
```

```python
rules = association_rules(frequent_items, metric = 'confidence', min_threshold = 0.3)
rules
```

```python
rules.sort_values(by = 'lift' ,ascending=False).head(10)
```

```python
prd = input("Enter a product:")
prd_rule = rules[rules['antecedents'].apply(lambda x : prd in x)]
if not prd_rule.empty:


  print(f"Since you bought {prd}, ypu might also like")
  for i in prd_rule['consequents']:
    print(list(i)[0])
else:
  print("No recommendstion found")
```
